```python
import pandas as pd

# Calculate the correlation matrix
correlation_matrix = dailyact.corr()

# Select the top features (excluding 'Calories')
top_features = correlation_matrix.abs().nlargest(6, 'Calories')['Calories'].index[1:]

# Create a new DataFrame to store the correlation values
correlation_values_df = pd.DataFrame(columns=['Feature', 'Correlation'])

# Assign correlation values to the new DataFrame
for feature in top_features:
    correlation_value = correlation_matrix.loc['Calories', feature]
    correlation_values_df = correlation_values_df.append({'Feature': feature, 'Correlation': correlation_value}, ignore_index=True)

# Print the new DataFrame
correlation_values_df
```

```python

correlation_matrix = dailyact.corr()

# Select the top features (excluding 'Calories')
top_features = correlation_matrix.abs().nlargest(6, 'Calories')['Calories'].index[1:]

# Create a new DataFrame to store the correlation values
correlation_values_df = pd.DataFrame(columns=['Feature', 'Correlation'])

# Assign correlation values to the new DataFrame
for feature in top_features:
    correlation_value = correlation_matrix.loc['Calories', feature]
    correlation_values_df.loc[len(correlation_values_df)] = [feature, correlation_value]

# Print the new DataFrame
correlation_values_df
```

This forms a dataframe consisting of top 5 correlated features among the entire dataset with respect to the Calories column


```python
top_features
```

```python
for feature in top_features:
    fix, ax = plt.subplots()
    ax.scatter(dailyact[feature], dailyact['Calories'])
    ax.set_xlabel(feature)
    ax.set_ylabel('Calories')
    ax.set_title(f'Scatter plot of {feature} vs Calories')
```

This is a depiction of how the values are distributed for each feature , now lets fit a trend line and understand the distribution more clearly 

```python
import seaborn as sns


# Scatter plots for top correlated features with trend line
for feature in top_features:
    sns.lmplot(x=feature, y='Calories', data=dailyact)
    plt.title(f"Scatter plot with Trend Line: {feature} vs Calories")
    plt.show()
```

```python

# Calculate z-scores for each column (excluding datetime column)
numerical_columns = dailyact.select_dtypes(include=np.number).columns
z_scores = (dailyact[numerical_columns] - dailyact[numerical_columns].mean()) / dailyact[numerical_columns].std()

# Define threshold for outliers
threshold = 3

# Find outliers for each column
outliers = (z_scores > threshold) | (z_scores < -threshold)

# Print the indices of outliers
for column in outliers.columns:
    column_outliers = outliers[column]
    if any(column_outliers):
        print(f"Outliers in column '{column}':")
        print(dailyact.index[column_outliers])
        print()

```

```python
outliers
```

```python
dailyact_no_outliers = dailyact[~outliers.any(axis=1)]
dailyact_no_outliers
```

This is the dataset now without outliers

```python
import seaborn as sns


# Scatter plots for top correlated features with trend line
for feature in top_features:
    sns.lmplot(x=feature, y='Calories', data=dailyact_no_outliers)
    plt.title(f"Scatter plot with Trend Line: {feature} vs Calories")
    plt.show()
```

```python



numeric_columns = dailyact.select_dtypes(include=['int64', 'float64']).columns

# Create subplots for each column
fig, axes = plt.subplots(nrows=len(numeric_columns), ncols=1, figsize=(8, 6*len(numeric_columns)))

# Create KDE plot for each column
for i, column in enumerate(numeric_columns):
    sns.kdeplot(data=dailyact, x=column, ax=axes[i])
    axes[i].set_xlabel(column)
    axes[i].set_ylabel('Density')

# Adjust spacing between subplots
fig.tight_layout()

# Show the plot
plt.show()
```
